```python
def cap_ba_bd(row):
    if row['bedrooms'] > 11:
        row['bedrooms'] = 3 
    if row['bedrooms'] > 9:
        row['bedrooms'] = 10
    if row['bathrooms'] < 1:
        row['bathrooms'] = 1
    return row
```

```python
df_capped = df.apply(cap_ba_bd, axis = 1)
```

# Exploring Some Features In The Data

Here we will look at some of the features houses have and see if there are satistically significant and see if they should be used in out model.

# Basement vs No Basement

```python
kc_mean_price = df.price.mean()
```

```python
basement = df[df['basement'] == 1]
basement_mean = basement.price.mean()

no_basement = df[df['basement'] == 0]
no_basement_mean = no_basement.price.mean()

label_basement = ["Basement", "No basement"]
values_basement = [basement_mean, no_basement_mean]
```

```python
basement_mean, no_basement_mean, kc_mean_price
```

```python
plt.subplots(figsize=(15,10))

sns.barplot(x=label_basement, y=values_basement, palette="rocket")

plt.ylabel("Average Prices ($)", size=20)
plt.title('Basement Affect on Price', size = 18)
plt.savefig("images/high_price_basement.png",bbox_inches = "tight")
plt.show()
```

```python
# Null: mu1 = mu2
# Alt: m1 != mu2

alpha = 0.05
basement_p_val = stats.ttest_ind(basement.price, no_basement.price, equal_var=False)[1]
print("Basement vs No Basement T-test P Value: ", basement_p_val)
if basement_p_val < 0.05:    
    print("Having a basement vs not having a basement has a statistically significant impact on average property value")
else:
    print("we are accepting null hypothesis")
```

# Top 5 Zipcodes

```python
kc_top5_price = df.groupby("zipcode")["price"].mean().sort_values(ascending = False)[:5]
area_labels = ["Medina", "Bellevue", "Mercer Island", 
               "Madison Park", "Capitol Hill"]
```

```python
plt.subplots(figsize=(15,10))
sns.barplot(x=kc_top5_price.index, y=kc_top5_price, order=kc_top5_price.index, palette="rocket") 
plt.xticks(np.arange(5),area_labels, rotation=75, size=8) 
plt.hlines(kc_mean_price, -.5 ,4.5, colors="blue", label="Average Price") 


plt.xlabel("Neighborhoods", size=20)
plt.ylabel("Prices ($1mil)", size=20)
plt.title("Neighborhoods with Highest Property Price", size=16, y=1.08)
plt.legend() 

plt.show();
```

```python
# Null: They are the same
# Alt: They are different

alpha = 0.05 
#ANOVA Test Setup
kc_top5 = df[df.zipcode.isin(kc_top5_price.index)]
formula = 'price~C(zipcode)'
lm_top5 = smf.ols(formula, kc_top5).fit()
anova_top5_table = sm.stats.anova_lm(lm_top5, typ=2)

if anova_top5_table["PR(>F)"][0] < alpha:
    print("Top 5 ranked zipcode have a statistically significant impact on average property value")
    print("Zipcode ANOVA F-statisic Probability: ", anova_top5_table["PR(>F)"][0])
```

# Feature Engineering

```python
# Creating new features of for sale quarter and revonvation age.
df["renovation_age"] = df.yr_renovated.apply(lambda x: 2020-x if x > 0 else 0)
df["sale_year"] = df.date.apply(lambda x: int(x[:4])) 
df["sale_quarter"] = df.date.apply(lambda x: int(x[4:6])//3.1 + 1) 
```

# Model Evaluation

```python
df_features = df.drop(['id', 'date', 'lat', 'long', 'price'], axis=1)
target = df['price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression
```

```python
def model_eval(features, target):
    lm = LinearRegression()
    lm.fit(features, target)
    print("R^2: ", lm.score(features, target))    
    
def model_score(features, target):
    X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=1,test_size=0.2)
    lm = linear_model.LinearRegression()
    lm.fit(X_train, y_train)
    y_train_pred = lm.predict(X_train)
    train_mae = metrics.mean_absolute_error(y_train, y_train_pred)
    train_mse = metrics.mean_squared_error(y_train, y_train_pred)
    train_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))
    price_std = target.std()
    
    y_pred = lm.predict(X_test)
    test_mae = metrics.mean_absolute_error(y_test, y_pred)
    test_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    
    selector = SelectKBest(f_regression, k=20)
    selector.fit(X_train, y_train)
    selector.get_support()
    
    print ("R^2 Score:", lm.score(X_train, y_train))
    
    print('Mean Absolute Error:', train_mae )
    print('Mean Squared Error:',  train_mse)
    print('Root Mean Squared Error:' , train_rmse)
    
    print('Z-Score of Mean Absolute Error:', train_mae/price_std )
    print('Z-Score of Root Mean Squared Error:' , train_rmse/price_std)
    
    print ("Test Score:", lm.score(X_test, y_test))
    print('Mean Absolute Error  Z:', test_mae/price_std )
    print('Root Mean Squared Error Z:' , test_rmse/price_std)
    print('Training: ', int(train_rmse), "vs. Testing: ", int(test_rmse))
```

```python
model_eval(df_features, target)
```
